````python
정규표현식(regex)
```
    1. 참조 : https://ko.wikipedia.org/wiki/%EC%A0%95%EA%B7%9C_%ED%91%9C%ED%98%84%EC%8B%9D
    2. 간략문법
        \d : (숫자와 매치. [0-9]와 동일)
        \D : (숫자가 아닌 문자)ㄴ
        \w : (영문자나 숫자). [a-zA-Z0-9]와 동일
        \W : (영문자나 숫자가 아닌 문자)   
        .  : (문자)
        \. : (.)
        \s : ( )
        {n,} : n회 반복
        {2,4} : 2~4회 반복
        ? : (0번이나 1번 의미)
        + : (1번 이상 반복)
        * : (0번 이상 반복)
    3. 정규표현식 연습장 : https://regexr.com/
        ex1. 전화번호 패턴 : 0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}
        ex2. 이메일 패턴 : \w[a-z0-9_]+@\w+(\.\w+){1,2}
```        
````

# 4절. 포맷팅
    - 포맷팅이란 출력할 문자, 숫자, 날짜 데이터에 형식을 지정하는 것
    - 변수와 텍스트를 결합해서 보기 좋은 형식을 새로운 텍스트를 만들 수 있음
## 4.1 문자열에 형식 지정

````python
name = "홍길동"
age = 30
print(name + '님 나이는' + str(age) + '살입니다')
print(name, '님 나이는', age, '살입니다', sep='')
````

````python
# %s문자, %d10진수정수, %ƒ실수, 
print("%s님 나이는 %d살입니다"%(name, age))
print(f"{name}님 나이는 {age}살입니다")
````

````python
print("{}님 나이는 {}살입니다".format(name, age))
print("{0}님 나이는 {1}살입니다".format(name, age))
print("{1}님 나이는 {0}살입니다".format(age, name))
print("{0}살 {1:*>4}님 반갑습니다. {1}님 어서오세요".format(age, name))
````

## 4.2 format()문법

- "{[인덱스]:[공백대체문자][정렬방식][자릿수][타입]}".format(변수명)
    * 인덱스 : format 함수 안의 매개변수 순서
    * 공백대체문자는 정렬방식과 자릿수를 명시하여야 함
    * 정렬방식 : <(왼쪽정렬), >(오른쪽정렬), ^(가운데정렬)
    * 정수자릿수 : 출력할 자리수
    * 실수자릿수 : 전체출력될자리수, 변수중출력할자리수
    * 타입 : d(10진수), f(실수 ex:4.1f:1은 소수점자리수), s(문자), o(8진수), x(16진수), b(2진수)

````python
name = input("이름은?")
age = int(input("나이?"))
````

````python
print("[{0:*^10d}]살 [{1:_>10}]님 안녕하세요".format(age, name))
print(f"[{age:*<10.1f}]살 [{name:_>10.2s}]님 안녕하세요")
````
