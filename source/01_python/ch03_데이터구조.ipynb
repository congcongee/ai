- index(찾을데이터) : 0번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환
- index(찾을데이터, start) : start번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
```

## 3.2 딕셔너리 item 수정, 추가, 제거

```python
my_favorite = {'fruit':'apple', 'number':12, 'sports':'yoga'}
# 한 item씩 추가
my_favorite['people'] = '홍길동'
print('1. 한 item 추가후 :', my_favorite)
# 여러 item을 한꺼번에 추가
my_favorite.update({'subject':'수학', 'major':'AI'})
print('2. 여러 item 추가 후 :', my_favorite)

# 한 item씩 수정
my_favorite['fruit'] = 'banana'
print('3. 한 item 수정 후 :', my_favorite)
# 여러 item을 한꺼번에 수정
# my_favorite.update({'fruit':'kiwi', 'number':-99})
my_favorite.update(fruit='kiwi', number=-99)
print('4. 여러 item 수정 후:', my_favorite)

# item 제거 방법1 : del 변수['삭제할key'] - 없는 key일 경우 에러
del my_favorite['fruit']
if 'fruit' in my_favorite:
    del my_favorite['fruit']
else:
    print('5. fruit 삭제하여 존재하지 않음')
# item 제거 방법2 : 변수.pop('삭제할key') - 해당 key값을 반환하고 제거 (없는 key일 경우 에러)
print('6. 제거한 데이터 :', my_favorite.pop('number'))
if 'number' in my_favorite:
    print('7. 제거한 데이터 :', my_favorite.pop('number'))
else:
    print('7. number 삭제하여 존재하지 않음')
print(my_favorite)
```

## 3.3 출력하기

- 리스트, 튜플, 셋을 for문에 사용시

```python
list(enumerate(names))
```

```python
for no, name in enumerate(names):
    print("{} : {}". format(no, name))
```

```python
names = ['길동', '길자', '순자']
tels  = ['010-9999-9999',
        '010-2222-2222',
        '010-7777-7777']
adds = ['서울', '인천', '판교']
for idx in range(len(names)): # [0,1,2]
    print("{}. {}의 연락처는 {}이고, {} 살아요".format(idx+1,
                                            names[idx],
                                            tels[idx],
                                            adds[idx]))
```

```python
list(zip(names, tels, adds))
```

```python
for name, tel, add in zip(names, tels, adds):
    print(f"{name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
list(enumerate(zip(names, tels, adds)))
```

```python
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    print(f"{idx+1}. {name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
# names, tels, adds를 딕셔너리 리스트로
members = []
for name, tel, add in zip(names, tels, adds):
    member = {'name':name, 'tel':tel, 'address':add}
    members.append(member)
members
```

```python
members= []
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    member = {'순위':idx+1, '이름':name, '전화':tel, '주소':add}
    members.append(member)
members
```
